```python
import matplotlib.pyplot as plt
import numpy as np

# Define the angles and corresponding distances (your data)
angles = np.array([25, 35, 45, 55, 65])
distances = np.array([
    [0.861, 0.868, 0.873, 0.884, 0.889, 0.861, 0.874, 0.88, 0.893, 0.878],
    [1.102, 1.109, 1.108, 1.104, 1.103, 1.102, 1.094, 1.094, 1.086, 1.081],
    [1.16, 1.157, 1.16, 1.161, 1.162, 1.164, 1.165, 1.167, 1.168, 1.156],
    [1.054, 1.052, 1.053, 1.055, 1.056, 1.058, 1.058, 1.059, 1.062, 1.054],
    [0.836, 0.84, 0.844, 0.845, 0.849, 0.849, 0.85, 0.853, 0.855, 0.856]
])

# Calculate the mean for each angle
mean_distances = np.mean(distances, axis=1)

# Fit a quadratic polynomial (degree 2) to the data
coefficients = np.polyfit(angles, mean_distances, 2)
polynomial = np.poly1d(coefficients)

# Generate angles for the fitted curve
fit_angles = np.linspace(25, 65, 500)
fit_distances = polynomial(fit_angles)

# Create the scatter plot with error bars
plt.errorbar(angles, mean_distances, yerr=np.std(distances, axis=1), fmt='o', color='red', label='Average Distance with Error Bars')
plt.scatter(angles, distances.flatten(), color='blue', alpha=0.5, label='Measured Data')

# Plot the fitted quadratic curve
plt.plot(fit_angles, fit_distances, color='green', linestyle='--', label='Fitted Parabolic Curve')

# Add labels and title
plt.xlabel('Angle (degrees)')
plt.ylabel('Distance (m)')
plt.title('Range vs Angle Graph')
plt.legend()

# Show the graph
plt.grid(True)
plt.show()

```
