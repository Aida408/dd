```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['price'], axis=1)
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
```

375512.jpg

38_blog_image_1.png

```python

import sklearn.linear_model as lm

lin_reg_model = lm.LinearRegression()
lin_reg_model.fit(X_train, y_train)
```

```python

lin_reg_model.intercept_
```

```python

lin_reg_model.coef_
```

```python
lin_reg_model.feature_names_in_
```

```python
print("  R**2   : {:.2%}".format(lin_reg_model.score(X_train, y_train)))
print("  R**2   : {:.2%}".format(lin_reg_model.score(X_test, y_test)))
```

```python

from sklearn.metrics import mean_squared_error


y_train_pred = lin_reg_model.predict(X_train)
y_test_pred  = lin_reg_model.predict(X_test)


train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
print("Train RMSE: ", train_rmse)

test_rmse = mean_squared_error(y_test, y_test_pred, squared=False)
print("Test RMSE: ", test_rmse)
```

```python

sns.scatterplot(x=y_test, y=y_test_pred, color="blue")
plt.title('Real vs Predicted test values for Linear Regression')
plt.xlabel('Real')
plt.ylabel('Predicted')
plt.show()
```

```python

import sklearn.tree as st

tree_reg_model = st.DecisionTreeRegressor()
tree_reg_model.fit(X_train, y_train)
```

```python

print("  R**2   : {:.2%}".format(tree_reg_model.score(X_train, y_train)))
print("  R**2   : {:.2%}".format(tree_reg_model.score(X_test, y_test)))
```

```python

from sklearn.metrics import mean_squared_error

y_train_pred = tree_reg_model.predict(X_train)
y_test_pred2  = tree_reg_model.predict(X_test)

train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
print("Train RMSE: ", train_rmse)

test_rmse = mean_squared_error(y_test, y_test_pred2, squared=False)
print("Test RMSE: ", test_rmse)

```
